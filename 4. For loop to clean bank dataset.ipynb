```python
bank = pd.DataFrame()

for i in np.arange(1,28):
    df = pd.read_excel('BANKS2020.xlsx', sheet_name=i)
    df = df[1:]
    df = df.T
    df.dropna(how='all', axis=1, inplace=True)
    df.dropna(how='all', axis=0, inplace=True)
    df[1][0] = 'year'
    df1 = df.copy()
    df1.columns = df1.iloc[0]
    df1 = df1[1:]
    df1['Symbol'] = df.index[0]
    df1.set_index('Symbol', inplace=True)
    df1.dropna(how='all', axis=1, inplace=True)
    df1.dropna(how='all', axis=0, inplace=True)
    df1 = df1.loc[:,df1.columns.notnull()]
    bank = pd.concat([bank, df1])
```
